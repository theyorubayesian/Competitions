# Facts

* Every galaxy has a certain limited potential for improvement in the index described by the following function:

```python
potential_for_index_increase = -np.log(index + 0.01) + 3

```

* Likely index increase dependent on potential for improvement and on extra energy availability is described by the following function:

```python
likely_increase_in_index = (extra_energy * potential_for_index_increase**2) / 1000

```

* The following constraints also exist:
    1. Total energy available for allocation is 50000 zillion DSML
    2. No galaxy should have more than 100 zillion DSML
    3. Galaxies with existence expectancy index below 0.7 should have 
        more than 10% of total energy to be allocated. 

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
````

````python
index_range = np.linspace(0, 0.8, num=100)
````

````python
potential_for_index_increase = -np.log(index_range + 0.01) + 3
````

````python
sns.lineplot(x=index_range, y=potential_for_index_increase)
````

````python
extra_energy = np.linspace(0, 100, num=100)
````

````python
from random import sample
````

````python
def plot_likely_index_increase_for_one_potential(index_potential=None):
    ind_pot = index_potential or sample(list(potential_for_index_increase), 1)[0]
    energy_range = np.linspace(0, 100, num=100)
    likely_index_increase = list(map(lambda x: x * ind_pot**2 / 1000, energy_range))
    print("Potential for index increase used is: ", ind_pot)
    sns.lineplot(x=energy_range, y=likely_index_increase)
    plt.show()
````

````python
for x in [3, 4, 5, 6, 7, 7.5]:
    plot_likely_index_increase_for_one_potential(index_potential=x)
````
